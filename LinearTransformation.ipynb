# Linear Transformation Example

Note: This is not a machine learning model itself but a preprocessing step commonly used in machine learning where to train the model, images are gets transformed to grey scale.

To generate a grey scale image
- get input (normal color image - selfie, photo)
- get its pixelated value (Each image consists of pixels arranged in a matrix format with rows and columns. Each pixel has Red (R), Green (G), and Blue (B) values.)
- apply transformation by using a formula (conversion to greyscale)
- display greyscale image

https://en.wikipedia.org/wiki/Grayscale

```python
import numpy as np # library to work with matrices
from skimage import io # library to work with images

image = io.imread('https://picsum.photos/seed/picsum/200/300') # reads the image as row x columns rgb values
height, width, channels  = image.shape

print(f"Rows (height): {height}") # how many rows are in the picture
print(f"Columns (width): {width}") # how many cols are in the picture
# print(image) # matrices of RGB values of the image will be printed (0-255)
print(image[:5, :5])  # First 5 rows, first 5 columns

print("Red Values", image[:5, :5, 0]) # Red Values of first 5 x 5 (row x col)
print("Green Values", image[:5, :5, 1]) # Green Values of first 5 x 5 (row x col)
print("Blue Values", image[:5, :5, 2]) # Blue Values of first 5 x 5 (row x col)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
plt.axis('off')
plt.show()
```

Now, Apply Transformation on the Matrix..
https://wikimedia.org/api/rest_v1/media/math/render/svg/c3a60e9ad9b8a5e5d2509cb9fc67dfef945d0038


```python
# Apply formula Y = 0.2126r + 0.7152g + 0.0722b (which is linear transformation)

import numpy as np
from skimage import io

# Apply the formula to create a grayscale image
# image [R G B]
# apply formula (ie) multiple all the Red values by 0.2126 + multiply all the Green Values by 0.7152 + multiply all the Blue Values by 0.0722
grayscale_image = 0.2126 * image[:, :, 0] + 0.7152 * image[:, :, 1] + 0.0722 * image[:, :, 2]
# For the image above, take the first R G B Values and create Grey Scale
# R = 219
# G = 186
# B = 193

# Y = (0.2126 × 219) + (0.7152 × 186) + (0.0722 × 193)
#  = 46.5594 + 133.0272 + 13.9346
#  = 193.5212 (Matches with the Output 193.5212 on the result below)
# instead of representing a pixel with 3 values - Red Green Blue value, we are representing it with only on grey scale (Linear Transformation)

print(grayscale_image[:5, :5])  # First 5 rows, first 5 columns

import matplotlib.pyplot as plt

plt.imshow(grayscale_image, cmap='grey')
plt.axis('off')
plt.show()
```
